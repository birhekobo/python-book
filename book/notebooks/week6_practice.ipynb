## Section 2: Instance vs Class Attributes & Methods

Class attributes are shared by all instances. Use `@classmethod` and `@staticmethod`.

```python
# Example: class attributes and methods
class Circle:
    pi = 3.14159  # Class attribute

    def __init__(self, radius):
        self.radius = radius  # Instance attribute

    def area(self):
        return Circle.pi * self.radius ** 2

    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)

    @staticmethod
    def description():
        return "A Circle represents a round shape."

c1 = Circle(5)
c2 = Circle.from_diameter(10)
print(f"Area: {c1.area():.2f}")
print(f"From diameter area: {c2.area():.2f}")
print(Circle.description())
```

```python
# Exercise 2: BankAccount with class attribute for interest rate
# TODO: Add class attribute interest_rate and a classmethod to change it
class BankAccount:
    interest_rate = 0.03  # Class attribute

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def apply_interest(self):
        self.balance += self.balance * BankAccount.interest_rate

    @classmethod
    def set_interest_rate(cls, rate):
        cls.interest_rate = rate

    def __str__(self):
        return f"{self.owner}: ${self.balance:.2f}"

acc = BankAccount("Alice", 1000)
acc.apply_interest()
print(acc)  # $1030.00
BankAccount.set_interest_rate(0.05)
acc.apply_interest()
print(acc)  # $1081.50
```

## Section 3: Magic / Dunder Methods

Special methods like `__str__`, `__repr__`, `__len__`, `__add__` customize behavior.

```python
# Example: magic methods
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):
        return int((self.x ** 2 + self.y ** 2) ** 0.5)

v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = v1 + v2
print(v3)       # Vector(4, 6)
print(len(v1))  # 5
```

## Section 4: Inheritance

A child class inherits attributes and methods from a parent class.

```python
# Example: inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog("Rex"), Cat("Whiskers"), Dog("Buddy")]
for a in animals:
    print(f"{a.name} says {a.speak()}")
```

## Section 5: Polymorphism

Polymorphism lets different classes define methods with the same name.

```python
# Example: polymorphism with duck typing
class Pigeon:
    def fly(self):
        return "Pigeon flies slowly."

class Airplane:
    def fly(self):
        return "Airplane flies fast!"

class Rocket:
    def fly(self):
        return "Rocket flies to space!"

def make_it_fly(flying_thing):
    print(flying_thing.fly())

make_it_fly(Pigeon())
make_it_fly(Airplane())
make_it_fly(Rocket())
```

## Section 6: super() and Method Resolution Order

Use `super()` to call parent class methods.

```python
# Example: using super()
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def work(self):
        return f"{self.name} is working."

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def work(self):
        return f"{self.name} is managing {self.department}."

mgr = Manager("Alice", 80000, "Engineering")
print(mgr.work())
print(f"Salary: ${mgr.salary}")
```

## Section 7: Composition vs Inheritance

Composition uses "has-a" relationships (an object contains other objects).

```python
# Exercise 7: Composition — a Library has Books
# TODO: Complete the Library class
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"'{self.title}' by {self.author}"

class Library:
    def __init__(self, name):
        self.name = name
        self.books = []  # Composition: Library has Books

    def add_book(self, book):
        self.books.append(book)

    def list_books(self):
        if not self.books:
            return "No books in library."
        return "\n".join(str(b) for b in self.books)

# Test the Library
lib = Library("City Library")
lib.add_book(Book("1984", "George Orwell"))
lib.add_book(Book("Dune", "Frank Herbert"))
print(f"{lib.name}:")
print(lib.list_books())
```

```python
# Exercise 8: Property decorators
# TODO: Use @property to create getters/setters for a Temperature class
class Temperature:
    def __init__(self, celsius=0):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value

    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32

    def __str__(self):
        return f"{self._celsius:.1f}°C / {self.fahrenheit:.1f}°F"

t = Temperature(25)
print(t)           # 25.0°C / 77.0°F
t.celsius = 100
print(t)           # 100.0°C / 212.0°F
print(t.fahrenheit)  # 212.0
```

```python
# Exercise 10: Abstract base class
# TODO: Create an abstract Employee class with an abstract method calculate_pay()
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    @abstractmethod
    def calculate_pay(self):
        pass

    def __str__(self):
        return f"{self.name} (ID: {self.emp_id})"

class SalariedEmployee(Employee):
    def __init__(self, name, emp_id, annual_salary):
        super().__init__(name, emp_id)
        self.annual_salary = annual_salary

    def calculate_pay(self):
        return self.annual_salary / 26  # Bi-weekly pay

class HourlyEmployee(Employee):
    def __init__(self, name, emp_id, hourly_rate, hours_worked):
        super().__init__(name, emp_id)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_pay(self):
        return self.hourly_rate * self.hours_worked

# Test
employees = [
    SalariedEmployee("Alice", 101, 78000),
    HourlyEmployee("Bob", 102, 35, 80)
]
for emp in employees:
    print(f"{emp}: ${emp.calculate_pay():.2f}")
```

## Section 9: Operator Overloading & Comparison Methods

```python
# Exercise 11: Overload comparison operators for a Money class
# TODO: Implement __lt__, __eq__, __add__ for Money
class Money:
    def __init__(self, amount, currency="USD"):
        self.amount = amount
        self.currency = currency

    def __str__(self):
        return f"{self.currency} {self.amount:.2f}"

    def __add__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot add different currencies")
        return Money(self.amount + other.amount, self.currency)

    def __lt__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot compare different currencies")
        return self.amount < other.amount

    def __eq__(self, other):
        return self.amount == other.amount and self.currency == other.currency

m1 = Money(100)
m2 = Money(50)
m3 = Money(100)
print(m1 + m2)         # USD 150.00
print(m1 < m2)         # False
print(m1 == m3)        # True
print(m1 > m2)         # True (uses __lt__ under the hood)
```

```python
# Exercise 12: Singleton pattern using class method
# TODO: Create a Logger class that only allows one instance
class Logger:
    _instance = None

    def __init__(self):
        self.logs = []

    @classmethod
    def get_instance(cls):
        if cls._instance is None:
            cls._instance = Logger()
        return cls._instance

    def log(self, message):
        self.logs.append(message)
        print(f"LOG: {message}")

    def show_logs(self):
        return "\n".join(self.logs)

# Test singleton
logger1 = Logger.get_instance()
logger2 = Logger.get_instance()
logger1.log("Application started")
logger2.log("User logged in")
print(f"Same instance: {logger1 is logger2}")
print(f"Total logs: {len(logger1.logs)}")
```

```python
# Exercise 13: Custom container class with __getitem__ and __len__
# TODO: Complete the ShoppingCart class
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price, quantity=1):
        self.items.append({"name": name, "price": price, "quantity": quantity})

    def __len__(self):
        return sum(item["quantity"] for item in self.items)

    def __getitem__(self, index):
        return self.items[index]

    def __iter__(self):
        return iter(self.items)

    def total(self):
        return sum(item["price"] * item["quantity"] for item in self.items)

    def __str__(self):
        lines = [f"{item['name']} x{item['quantity']} - ${item['price']*item['quantity']:.2f}" for item in self.items]
        return "\n".join(lines) + f"\nTotal: ${self.total():.2f}"

cart = ShoppingCart()
cart.add_item("Apple", 0.50, 6)
cart.add_item("Bread", 2.99, 2)
cart.add_item("Milk", 3.49, 1)
print(cart)
print(f"\nFirst item: {cart[0]}")
print(f"Total items: {len(cart)}")
```
