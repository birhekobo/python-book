## Section 1: Lists

## Section 2: List Comprehensions

A concise way to create lists: `[expression for item in iterable if condition]`

```python
# Example: list comprehensions
squares = [x**2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
print(f"Squares: {squares}")
print(f"Evens: {evens}")
```

```python
# Exercise 3: Use a list comprehension to get odd numbers 1-50
# TODO: Write a comprehension for odd numbers
odds = [x for x in range(1, 51) if x % 2 != 0]
print(odds)
```

```python
# Exercise 4: Convert temperatures using comprehension
# TODO: Convert [32, 68, 86, 104] Fahrenheit to Celsius using a list comprehension
# Formula: C = (F - 32) * 5/9
f_temps = [32, 68, 86, 104]
c_temps = [round((f - 32) * 5/9, 1) for f in f_temps]
print(c_temps)  # Expected: [0.0, 20.0, 30.0, 40.0]
```

## Section 3: Tuples

Tuples are ordered, **immutable** sequences. Good for fixed data.

```python
# Example: tuple operations
point = (3, 4)
x, y = point  # unpacking
rgb = (255, 128, 0)
print(f"Point: {point}, x={x}, y={y}")
print(f"RGB red component: {rgb[0]}")
print(f"Count of 255: {rgb.count(255)}")
```

## Section 4: Dictionaries

Dictionaries store key-value pairs. Keys must be immutable.

```python
# Example: dictionary operations
student = {
    "name": "Alice",
    "age": 22,
    "courses": ["Math", "Physics"],
    "gpa": 3.8
}
print(f"Name: {student['name']}")
print(f"Keys: {student.keys()}")
print(f"Values: {student.values()}")
student["graduated"] = False  # add new key
print(student.get("major", "Undeclared"))  # safe get with default
```

## Section 5: Dictionary Comprehensions

Like list comprehensions but for dicts: `{key: value for item in iterable}`

```python
# Example: dict comprehension
squares_dict = {x: x**2 for x in range(1, 6)}
print(squares_dict)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 5}
```

```python
# Exercise 8: Create a dictionary mapping word to its length
# TODO: Use a dict comprehension
animals = ["dog", "elephant", "cat", "giraffe"]
lengths = {animal: len(animal) for animal in animals}
print(lengths)  # Expected: {'dog': 3, 'elephant': 8, 'cat': 3, 'giraffe': 7}
```

## Section 6: Sets

Sets are unordered collections of **unique** elements. Support mathematical operations.

```python
# Example: set operations
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
print(f"Union: {a | b}")
print(f"Intersection: {a & b}")
print(f"Difference (a-b): {a - b}")
print(f"Symmetric diff: {a ^ b}")
a.add(6)
a.discard(1)
print(f"After add/discard: {a}")
```

```python
# Exercise 9: Remove duplicates from a list using a set
# TODO: Convert list to set and back to get unique elements
numbers = [1, 2, 2, 3, 4, 4, 5, 1, 3]
unique = list(set(numbers))
print(unique)  # Expected: [1, 2, 3, 4, 5] (order may vary)
```

## Section 7: Set Comprehensions & Practical Use

Set comprehensions use `{expr for item in iterable}` syntax.

```python
# Example: set comprehension
even_squares = {x**2 for x in range(10) if x % 2 == 0}
print(even_squares)  # {0, 4, 16, 36, 64}
```

```python
# Exercise 14: Matrix transposition with nested lists
# TODO: Transpose the given 3x3 matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
# Transpose using nested list comprehension
transposed = [[row[i] for row in matrix] for i in range(3)]
print(transposed)  # Expected: [[1, 4, 7], [2, 5, 8], [3, 6, 9]]
```
