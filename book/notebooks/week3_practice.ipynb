## Section 2: Arguments & Parameters

Functions can have default arguments, keyword arguments, and variable-length args.

```python
# Example: default and keyword arguments
def power(base, exp=2):
    """Raise base to exponent exp (default 2)."""
    return base ** exp

print(power(5))      # 5^2 = 25
print(power(2, 10))  # 2^10 = 1024
print(power(exp=3, base=4))  # keyword args
```

```python
# Example: *args and **kwargs
def summarize(*args, **kwargs):
    print("Positional:", args)
    print("Keyword:", kwargs)

summarize(1, 2, 3, name="Alice", age=30)
```

## Section 4: Lambda Functions

Lambdas are small anonymous functions defined with `lambda args: expression`.

```python
# Example: lambda with map, filter, and sorted
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
evens = list(filter(lambda x: x % 2 == 0, nums))
print(f"Squared: {squared}")
print(f"Evens: {evens}")

students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
students.sort(key=lambda s: s[1], reverse=True)
print(f"Sorted by grade: {students}")
```

```python
# Exercise 6: Use lambda with map to convert Celsius to Fahrenheit
# TODO: Convert [0, 10, 20, 30, 40] Celsius to Fahrenheit using map + lambda
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print(fahrenheit)  # Expected: [32.0, 50.0, 68.0, 86.0, 104.0]
```

## Section 5: Recursion

A recursive function calls itself. Always include a base case to stop.

```python
# Example: factorial recursively
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

for i in range(6):
    print(f"{i}! = {factorial(i)}")
```

## Section 6: Scope & Docstrings

Variables defined inside a function are local. Use docstrings to document functions.

```python
# Example: local vs global scope
x = 10  # global

def modify():
    x = 20  # local — does not affect global
    print(f"Inside: {x}")

modify()
print(f"Outside: {x}")
```
