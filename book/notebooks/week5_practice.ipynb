## Section 3: Working with CSV

CSV (Comma-Separated Values) is a common data format. Use the `csv` module.

```python
# Example: writing and reading CSV
import csv

data = [
    ["Name", "Age", "City"],
    ["Alice", 25, "New York"],
    ["Bob", 30, "London"],
    ["Charlie", 35, "Tokyo"]
]

with open("people.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

with open("people.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

os.remove("people.csv")
```

## Section 5: Exception Handling

Use `try`/`except`/`else`/`finally` to handle errors gracefully.

```python
# Example: try-except-else-finally
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
    except TypeError:
        print("Error: Both arguments must be numbers!")
        return None
    else:
        print(f"Division successful: {result}")
        return result
    finally:
        print("Division attempt completed.")

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "a"))
```

## Section 6: Custom Exceptions

You can define your own exception classes by inheriting from `Exception`.

```python
# Exercise 9: Custom exception for age validation
# TODO: Define AgeError, write a function that raises it if age < 0 or age > 150
class AgeError(Exception):
    pass

def set_age(age):
    if age < 0 or age > 150:
        raise AgeError(f"Invalid age: {age}. Must be between 0 and 150.")
    return f"Age set to {age}"

# Test with try/except
try:
    print(set_age(200))
except AgeError as e:
    print(f"Caught: {e}")
```

## Section 8: Context Managers & the `with` Statement

The `with` statement simplifies resource management. You can also create custom context managers.

```python
# Exercise 13: Custom context manager using a class
# TODO: Complete the FileLogger context manager
class FileLogger:
    """Context manager that logs file operations."""
    def __init__(self, filename, mode="r"):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        print(f"Opening {self.filename} in mode '{self.mode}'")
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Closing {self.filename}")
        self.file.close()
        if exc_type:
            print(f"Exception occurred: {exc_val}")
        return False  # Don't suppress exceptions

# Test the custom context manager
with FileLogger("test_log.txt", "w") as f:
    f.write("Testing context manager\n")

with FileLogger("test_log.txt", "r") as f:
    print(f.read())

os.remove("test_log.txt")
```

```python
# Exercise 14: Reading a CSV and computing column statistics
# TODO: Read a CSV of weather data and compute average temperature
import csv

weather_data = [
    ["day", "temp", "humidity"],
    ["Monday", 25, 60],
    ["Tuesday", 28, 55],
    ["Wednesday", 22, 70],
    ["Thursday", 26, 65],
    ["Friday", 30, 50]
]

with open("weather.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(weather_data)

temperatures = []
with open("weather.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        temperatures.append(int(row["temp"]))

avg_temp = sum(temperatures) / len(temperatures)
print(f"Average temperature: {avg_temp:.1f}°C")
print(f"Max temperature: {max(temperatures)}°C")
print(f"Min temperature: {min(temperatures)}°C")

os.remove("weather.csv")
```
