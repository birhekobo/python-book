## Section 5: Strings

Strings support indexing, slicing, concatenation, and many methods.

```python
# Example: string operations
greeting = "Hello, Python!"
print(greeting[0])        # First character
print(greeting[-1])       # Last character
print(greeting[7:13])     # Slice
print(greeting.upper())   # Uppercase
print(greeting.replace("Python", "World"))
```

## Section 6: Arithmetic Operators

Python supports `+`, `-`, `*`, `/`, `//`, `%`, `**`.

```python
# Example: arithmetic operations
a, b = 15, 4
print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Floor division: {a // b}")
print(f"Modulus: {a % b}")
print(f"Power: {a ** b}")
```
