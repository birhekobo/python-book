## Step 1: Define the Transaction Class

Start by modeling a single transaction.

```python
# Step 1: Create the Transaction class
# It should store: amount (float), category (str), type ("income"/"expense"), description (str)
# Include a __str__ method for display and a to_dict method for JSON serialization

class Transaction:
    def __init__(self, amount, category, trans_type, description=""):
        self.amount = amount
        self.category = category
        self.trans_type = trans_type  # "income" or "expense"
        self.description = description

    def __str__(self):
        prefix = "+" if self.trans_type == "income" else "-"
        return f"{prefix}${self.amount:.2f} [{self.category}] {self.description}"

    def to_dict(self):
        return {
            "amount": self.amount,
            "category": self.category,
            "type": self.trans_type,
            "description": self.description
        }

# Quick test
t1 = Transaction(500, "Salary", "income", "Monthly salary")
t2 = Transaction(15.50, "Food", "expense", "Lunch")
print(t1)
print(t2)
print(t1.to_dict())
```

## Step 2: Build the FinanceTracker Class

This class manages all transactions with methods to add, list, compute balance, and persist data.

```python
# Step 2: Create the FinanceTracker class
# It holds a list of Transaction objects
# Methods: add_transaction, get_balance, get_summary, save_to_file, load_from_file

import json
import os

class FinanceTracker:
    def __init__(self, filename="finances.json"):
        self.transactions = []
        self.filename = filename

    def add_transaction(self, amount, category, trans_type, description=""):
        """Add a new transaction."""
        t = Transaction(amount, category, trans_type, description)
        self.transactions.append(t)
        print(f"Added: {t}")

    def get_balance(self):
        """Calculate net balance (income - expense)."""
        balance = 0.0
        for t in self.transactions:
            if t.trans_type == "income":
                balance += t.amount
            else:
                balance -= t.amount
        return balance

    def list_transactions(self):
        """Display all transactions."""
        if not self.transactions:
            print("No transactions recorded.")
            return
        print(f"\n{'='*50}")
        print("ALL TRANSACTIONS:")
        print(f"{'='*50}")
        for i, t in enumerate(self.transactions, 1):
            print(f"{i:3d}. {t}")
        print(f"{'='*50}")
        print(f"Net Balance: ${self.get_balance():.2f}")

    def get_summary(self):
        """Return a summary dict with total income, expense, and balance."""
        total_income = sum(t.amount for t in self.transactions if t.trans_type == "income")
        total_expense = sum(t.amount for t in self.transactions if t.trans_type == "expense")
        return {
            "total_income": total_income,
            "total_expense": total_expense,
            "balance": total_income - total_expense,
            "count": len(self.transactions)
        }

    def save_to_file(self):
        """Save all transactions to a JSON file."""
        data = [t.to_dict() for t in self.transactions]
        try:
            with open(self.filename, "w") as f:
                json.dump(data, f, indent=2)
            print(f"Saved {len(data)} transactions to {self.filename}")
        except IOError as e:
            print(f"Error saving file: {e}")

    def load_from_file(self):
        """Load transactions from a JSON file."""
        if not os.path.exists(self.filename):
            print(f"No file found: {self.filename}")
            return
        try:
            with open(self.filename, "r") as f:
                data = json.load(f)
            self.transactions = []
            for item in data:
                t = Transaction(
                    amount=item["amount"],
                    category=item["category"],
                    trans_type=item["type"],
                    description=item.get("description", "")
                )
                self.transactions.append(t)
            print(f"Loaded {len(self.transactions)} transactions from {self.filename}")
        except (json.JSONDecodeError, IOError, KeyError) as e:
            print(f"Error loading file: {e}")

# Quick test
tracker = FinanceTracker("test_finances.json")
tracker.add_transaction(3000, "Salary", "income", "June salary")
tracker.add_transaction(200, "Food", "expense", "Groceries")
tracker.add_transaction(150, "Utilities", "expense", "Electric bill")
tracker.list_transactions()
tracker.save_to_file()
tracker.load_from_file()
os.remove("test_finances.json")  # Clean up
```

```python
# Step 3: Add category analysis to FinanceTracker
# This method aggregates expenses by category

def expense_by_category(self):
    """Return a dict mapping category -> total expense."""
    category_totals = {}
    for t in self.transactions:
        if t.trans_type == "expense":
            category_totals[t.category] = category_totals.get(t.category, 0) + t.amount
    return category_totals

# Monkey-patch the method onto FinanceTracker for demonstration
FinanceTracker.expense_by_category = expense_by_category

# Test with the tracker
tracker = FinanceTracker()
tracker.add_transaction(5000, "Salary", "income")
tracker.add_transaction(1200, "Rent", "expense")
tracker.add_transaction(300, "Food", "expense")
tracker.add_transaction(150, "Transport", "expense")
tracker.add_transaction(400, "Food", "expense")
tracker.add_transaction(200, "Entertainment", "expense")

cats = tracker.expense_by_category()
print("Expenses by Category:")
for cat, total in sorted(cats.items(), key=lambda x: x[1], reverse=True):
    print(f"  {cat}: ${total:.2f}")
```

## Step 4c: Filtering & Searching with Loops

Use loops and conditionals to search through transaction history.

```python
# Step 4c: Search and filter transactions
# TODO: Implement search functionality using loops and string methods

def find_by_category(tracker, category):
    """Return all transactions matching a given category."""
    results = []
    for t in tracker.transactions:
        if t.category.lower() == category.lower():
            results.append(t)
    return results

def find_by_keyword(tracker, keyword):
    """Return all transactions whose description contains a keyword."""
    keyword = keyword.lower()
    results = []
    for t in tracker.transactions:
        if keyword in t.description.lower():
            results.append(t)
    return results

def summarize_month(tracker, income_threshold=0):
    """Categorize transactions as high/low value."""
    high_value = []
    low_value = []
    for t in tracker.transactions:
        if t.amount >= income_threshold:
            high_value.append(t)
        else:
            low_value.append(t)
    return high_value, low_value

# Test the search functions
tracker = FinanceTracker()
tracker.add_transaction(3000, "Salary", "income", "Monthly salary")
tracker.add_transaction(450, "Food", "expense", "Weekly groceries")
tracker.add_transaction(1200, "Rent", "expense", "Monthly apartment rent")
tracker.add_transaction(150, "Food", "expense", "Restaurant dinner")

print("\nFood transactions:")
for t in find_by_category(tracker, "Food"):
    print(f"  {t}")

high, low = summarize_month(tracker, 500)
print(f"\nHigh value (>= $500): {len(high)} transactions")
print(f"Low value (< $500): {len(low)} transactions")
```

## Step 5: Data Validation & Exception Handling

Add robust input validation to the tracker.

```python
# Step 5: Input validation helpers
# TODO: Implement get_positive_float and get_non_empty_string functions

def get_positive_float(prompt):
    """Ask user for a positive number, retry until valid."""
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Amount must be positive. Try again.")
                continue
            return value
        except ValueError:
            print("Invalid number. Please enter a numeric value.")

def get_non_empty_string(prompt):
    """Ask user for a non-empty string."""
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("Input cannot be empty. Try again.")

def get_yes_no(prompt):
    """Ask user a yes/no question, return True for yes."""
    while True:
        value = input(prompt + " (y/n): ").strip().lower()
        if value in ("y", "yes"):
            return True
        if value in ("n", "no"):
            return False
        print("Please answer 'y' or 'n'.")

# Test the validation functions
# amount = get_positive_float("Enter amount: ")
# cat = get_non_empty_string("Enter category: ")
# print(f"Validated: {amount}, {cat}")
print("Validation functions defined. Uncomment to test.")
```

## Step 6: Budget Alerts with Sets

Use sets to track budget categories and raise alerts.

```python
# Step 6: Budget alerts
# TODO: Complete the budget alert feature

class BudgetTracker:
    """Tracks spending limits per category."""
    def __init__(self):
        self.budgets = {}  # category -> limit

    def set_budget(self, category, limit):
        self.budgets[category] = limit
        print(f"Budget set: {category} = ${limit:.2f}")

    def check_alerts(self, expense_totals):
        """Check if any category exceeds its budget.
        expense_totals: dict of category -> total spent
        Returns list of alert messages."""
        alerts = []
        for category, spent in expense_totals.items():
            if category in self.budgets and spent > self.budgets[category]:
                alerts.append(
                    f"ALERT: {category} ${spent:.2f} exceeds budget ${self.budgets[category]:.2f}!"
                )
        return alerts

# Test budget alerts
bt = BudgetTracker()
bt.set_budget("Food", 500)
bt.set_budget("Entertainment", 200)

expenses = {"Food": 650, "Rent": 1200, "Entertainment": 50}
alerts = bt.check_alerts(expenses)
for alert in alerts:
    print(alert)

if not alerts:
    print("All categories within budget.")
```

## Step 7: Generate a Financial Report

Use comprehensions, string formatting, and file I/O to generate a report.

```python
# Step 7: Generate a financial report and save to a text file
# TODO: Complete the generate_report function

def generate_report(tracker, filename="report.txt"):
    """Generate a formatted financial report and save to file."""
    summary = tracker.get_summary()
    cats = tracker.expense_by_category()

    lines = []
    lines.append("=" * 60)
    lines.append("FINANCIAL REPORT")
    lines.append("=" * 60)
    lines.append(f"Total Income:    ${summary['total_income']:>8.2f}")
    lines.append(f"Total Expense:   ${summary['total_expense']:>8.2f}")
    lines.append(f"Net Balance:     ${summary['balance']:>8.2f}")
    lines.append(f"Transactions:    {summary['count']:>8}")
    lines.append("-" * 60)
    lines.append("Expenses by Category:")

    if cats:
        for cat, total in sorted(cats.items(), key=lambda x: x[1], reverse=True):
            lines.append(f"  {cat:20s} ${total:>8.2f}")
    else:
        lines.append("  No expenses recorded.")

    lines.append("=" * 60)

    report = "\n".join(lines)
    print(report)

    try:
        with open(filename, "w") as f:
            f.write(report)
        print(f"\nReport saved to {filename}")
    except IOError as e:
        print(f"Error saving report: {e}")

# Test report generation
tracker = FinanceTracker()
tracker.add_transaction(5000, "Salary", "income")
tracker.add_transaction(1200, "Rent", "expense")
tracker.add_transaction(350, "Food", "expense")
tracker.add_transaction(80, "Transport", "expense")
tracker.add_transaction(200, "Entertainment", "expense")
generate_report(tracker, "test_report.txt")
os.remove("test_report.txt")  # Clean up
```

## Step 8b: Visual Bar Chart with Strings

Use string repetition to create a simple text-based bar chart of expenses.

```python
# Step 8b: Text-based bar chart using string operations
# TODO: Complete the display_bar_chart function

def display_bar_chart(category_totals):
    """Display a simple text bar chart of expense categories."""
    if not category_totals:
        print("No data to display.")
        return

    max_val = max(category_totals.values())
    scale = 40 / max_val if max_val > 0 else 1

    print("\nEXPENSE BAR CHART:")
    print("-" * 50)
    for cat, total in sorted(category_totals.items(), key=lambda x: x[1], reverse=True):
        bar_length = int(total * scale)
        bar = "█" * bar_length
        print(f"{cat:15s} | {bar} ${total:.2f}")

# Test with sample data
sample_expenses = {"Rent": 1200, "Food": 450, "Transport": 150, "Entertainment": 200, "Utilities": 300}
display_bar_chart(sample_expenses)
```
