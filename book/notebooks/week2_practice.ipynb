## Section 1: Booleans & Comparison Operators

Booleans are `True` or `False`. Comparisons use `==`, `!=`, `<`, `>`, `<=`, `>=`.

```python
# Example: boolean expressions
a, b = 10, 20
print(f"a == b: {a == b}")
print(f"a != b: {a != b}")
print(f"a < b:  {a < b}")
print(f"a >= b: {a >= b}")
```

## Section 2: if / elif / else

Use conditionals to execute code only when a condition is met.

```python
# Example: if-elif-else chain
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

A `while` loop repeats as long as its condition is `True`.

```python
# Example: countdown with while
count = 5
while count > 0:
    print(count)
    count -= 1
print("Blast off!")
```

## Section 4: for Loops

## Section 5: Nested Loops

Loops inside loops are useful for working with 2D structures and printing patterns.

```python
# Example: nested loop printing a 3x3 grid
for row in range(3):
    for col in range(3):
        print(f"({row},{col})", end=" ")
    print()
```

## Section 6: break, continue, else in Loops

`break` exits the loop entirely. `continue` skips to the next iteration.
`else` on a loop runs if no `break` occurred.

```python
# Example: break and continue
for i in range(10):
    if i == 3:
        continue  # Skip 3
    if i == 7:
        break     # Stop at 7
    print(i, end=" ")
```
